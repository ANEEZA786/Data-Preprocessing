# **Outliers Removal with 'Percentile' Technique**

```
import numpy as np
import pandas as pd
```

```
df=pd.read_csv('/content/heights.csv')
print(df)
```

```
P=np.percentile(df.height,[5,95])
print(P)
```

```
new_df=df[(df['height']>P[0])&(df['height']<P[1])]
print(new_df)
```
